```python
df = pd.read_csv("allShades.csv")
df
```

```python
df.shape
```

```python
df.isna().sum()
```

```python
def convert_to_percentage(value):
    return value / 255
```

```python
# drop hue, sat, lightness
df = df.drop(["hue", "sat", "lightness"], axis=1)
```

```python
hexes = np.array([hex for hex in df["hex"]])
hsv = np.array([ImageColor.getcolor(hex, "HSV") for hex in hexes])
h = hsv[:, 0]
s = hsv[:, 1]
v = hsv[:, 2]

v = np.array([convert_to_percentage(value) for value in v])

# create a dataframe with the new columns
df["Hue"] = h
df["Saturation"] = s
df["Value"] = v

# save it to dataframe
df.to_csv("allShades_new.csv", index=False)
```
